```python
import os
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
%%time
def preprocess(i):
    try:
        df = pd.read_excel('./final/'+str(i),engine="openpyxl",skiprows=[0,1,2,3,4])
        df['Unnamed: 2'] = df['Unnamed: 2'].ffill()
        df = df.groupby(['Unnamed: 2','Unnamed: 3']).agg(sum).reset_index()
        df = df[(df['Unnamed: 2'].str.contains("Respiratory")) | (df['Unnamed: 2'].str.contains("weight"))]
        df.rename(columns={'Unnamed: 2': 'medical', 'Unnamed: 3': 'cat'}, inplace=True)
        df = df[[i for i in df.columns if i.startswith('U')==False]]
        df['year'] = str(i)[6:10]
        return df
    except:
        df = pd.read_excel('./final/'+str(i),engine="openpyxl",skiprows=[0,1,2,3,4])
        df = df.groupby(['Unnamed: 2','Unnamed: 3']).agg(sum).reset_index()
        df = df[(df['Unnamed: 2'].str.contains("Respiratory")) | (df['Unnamed: 2'].str.contains("weight"))]
        df = df.drop(columns=['Unnamed: 0','Unnamed: 1'])
        df.rename(columns={'Unnamed: 2': 'medical', 'Unnamed: 3': 'cat'}, inplace=True)
        df = df[[i for i in df.columns if i.startswith('U')==False]]
        return df

data = []
for i in os.listdir('./final'):
    data.append(preprocess(i))
data = pd.concat(data)

data = data[~data['medical'].str.contains("discharged")]
data['cat'] = data['cat'].replace({'4. Total': 'TOTAL'})
data = data[data['cat'].str.contains("TOTAL")]
data.loc[:,'category'] = data['medical'].str.contains("weight")
data['category'] = data['category'].map({True:'weight',False:'respiratory'})
data = data.drop(columns=['cat'])
data
```
